```python
for columna in df.columns:
    df[columna] = df[columna].replace([np.nan, pd.NA], None)
```

```python
def convertir_a_entero(valor):
    try:
        if pd.isnull(valor):
            return None
        else:
            return int(valor)
    except:
        return None
```

```python
df["id_marca"] = df["id_marca"].apply(convertir_a_entero).astype('Int64')
df["id_modelo"] = df["id_modelo"].apply(convertir_a_entero).astype('Int64')
df["mes_matriculacion"] = df["mes_matriculacion"].apply(convertir_a_entero).astype('Int64')
df["ano_matriculacion"] = df["ano_matriculacion"].apply(convertir_a_entero).astype('Int64')
df["id_distintivo_ambiental"] = df["id_distintivo_ambiental"].apply(convertir_a_entero).astype('Int64')
df["garantia"] = df["garantia"].apply(convertir_a_entero).astype('Int64')
df["id_concesionario"] = df["id_concesionario"].apply(convertir_a_entero).astype('Int64')
df["id_provincia"] = df["id_provincia"].apply(convertir_a_entero).astype('Int64')
df["id_traccion"] = df["id_traccion"].apply(convertir_a_entero).astype('Int64')
df["largo"] = df["largo"].apply(convertir_a_entero).astype('Int64')
df["ancho"] = df["ancho"].apply(convertir_a_entero).astype('Int64')
df["alto"] = df["alto"].apply(convertir_a_entero).astype('Int64')
df["capacidad_maletero"] = df["capacidad_maletero"].apply(convertir_a_entero).astype('Int64')
df["num_puertas"] = df["num_puertas"].apply(convertir_a_entero).astype('Int64')
df["num_plazas"] = df["num_plazas"].apply(convertir_a_entero).astype('Int64')
df["batalla"] = df["batalla"].apply(convertir_a_entero).astype('Int64')
df["peso"] = df["peso"].apply(convertir_a_entero).astype('Int64')
df["id_combustible"] = df["id_combustible"].apply(convertir_a_entero).astype('Int64')
df["num_cilindros"] = df["num_cilindros"].apply(convertir_a_entero).astype('Int64')
df["cilindrada"] = df["cilindrada"].apply(convertir_a_entero).astype('Int64')
df["id_sobrealimentacion"] = df["id_sobrealimentacion"].apply(convertir_a_entero).astype('Int64')
df["id_tipo_cambio"] = df["id_tipo_cambio"].apply(convertir_a_entero).astype('Int64')
df["num_marchas"] = df["num_marchas"].apply(convertir_a_entero).astype('Int64')
df["potencia_kw"] = df["potencia_kw"].apply(convertir_a_entero).astype('Int64')
df["potencia_cv"] = df["potencia_cv"].apply(convertir_a_entero).astype('Int64')
df["par"] = df["par"].apply(convertir_a_entero).astype('Int64')
df["velocidad_max"] = df["velocidad_max"].apply(convertir_a_entero).astype('Int64')
df["kilometraje"] = df["kilometraje"].apply(convertir_a_entero).astype('Int64')
```

```python
for columna in df.columns:
    df[columna] = df[columna].replace([np.nan, pd.NA], None)
```
